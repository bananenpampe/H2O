```python
import equistore
import rascaline
import rascaline_torch

```

```python
hypers_sr = {
    "cutoff": 5.0,
    "max_radial": 5,
    "max_angular": 5,
    "atomic_gaussian_width": 0.3,
    "center_atom_weight": 1.0,
    "radial_basis": {
        "Gto": {},
    },
    "cutoff_function": {
        "ShiftedCosine": {"width":0.5},
    },
    "radial_scaling":{"Willatt2018": {"exponent": 3.0, "rate": 1.5, "scale": 2.0}}
}

# --- define calculator ---
```

```python
class ReshapeModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.calculator = rascaline_torch.Calculator(rascaline.SoapPowerSpectrum(**hypers_sr))
        self.ModuleDict = torch.nn.ModuleDict({})

    def initialize_weights(self, positions, species, cell):

        system = rascaline_torch.System(species, positions, cell)

        feat = self.calculator(system)
        feat = feat.keys_to_properties(["species_neighbor_1","species_neighbor_2"])
        
        for key, block in feat.items():
            key = str(key)
            self.ModuleDict[key] = torch.nn.Linear(block.values.shape[1], 1)

    def forward(self, positions, species, cell):
        
        
        system = rascaline_torch.System( species, positions, cell)

        feat = self.calculator(system)

        feat = feat.keys_to_properties(["species_neighbor_1","species_neighbor_2"])

        out_blocks = []

        for key, block in feat.items():

            key = str(key)
            
            out_val = self.ModuleDict[key](block.values)

            out_block = equistore.TensorBlock(values=out_val,
                                              samples=block.samples,
                                              components=block.components,
                                              properties=equistore.Labels.range("energy", 
                                              out_val.shape[-1]))
            
            out_blocks.append(out_block)

        out_map = equistore.TensorMap(feat.keys, out_blocks)

        out_map = out_map.keys_to_samples(["species_center"])

        E = equistore.sum_over_samples(out_map, sample_names=["center","species_center"])
            
        return out_map
```

```python
def run_computation( positions, species, cell ):



    model = ReshapeModel()
    model.initialize_weights(positions, species, cell)
    
    return model.forward(positions, species, cell)
```
